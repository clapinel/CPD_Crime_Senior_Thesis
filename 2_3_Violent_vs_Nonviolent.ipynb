```
# make time increment buckets
# [00:00-3:59]
# [04:00-7:59]
# [08:00-11:59]
# [12:00-15:59]
# [16:00-19:59]
# [20:00-23:59]

def time_increments(row):
  if row.Date_24h.hour>0 and row.Date_24h.hour<4:
    val = "[00:00-3:59]"
  elif row.Date_24h.hour<8:
    val = "[04:00-7:59]"
  elif row.Date_24h.hour<12:
    val = "[08:00-11:59]"
  elif row.Date_24h.hour<16:
    val = "[12:00-15:59]"
  elif row.Date_24h.hour<20:
    val = "[16:00-19:59]"
  else:
    val = "[20:00-23:59]"
  return val 

all_results['Time_Bucket'] = all_results.apply(time_increments, axis=1)
```
