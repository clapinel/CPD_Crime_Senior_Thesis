```
import pandas as pd
```

```
file_name = '/content/drive/My Drive/Crimes_-_2001_to_present.csv'
```

```
df = pd.read_csv(file_name)
# filter to keep certain years 
df = df.query('Year>2007')
```

```
df.shape
```

```
na_rows = df[df.isna().any(axis=1)]
print(na_rows.shape)
na_rows
```

```
# df = df.apply(pd.to_numeric, errors='coerce')
df = df.dropna()
print(df.shape)
```

```
df['Primary Type'].value_counts()
```

```
types = df['Primary Type'].value_counts().keys().tolist()
print(len(types))
```

```
# clean the primary types 
# crim sexual assault and non-criminal 
df = df.replace('CRIMINAL SEXUAL ASSAULT','CRIM SEXUAL ASSAULT')
df = df.replace('NON - CRIMINAL','NON-CRIMINAL')
df = df.replace('NON-CRIMINAL (SUBJECT SPECIFIED)','NON-CRIMINAL')
```

```
df['Primary Type'].value_counts()
```

```
types = df['Primary Type'].value_counts().keys().tolist()
print(len(types))
```

```
# mark as violent or non-violent 
violent = ['HOMICIDE', 'ASSAULT', 'ROBBERY', 'CRIM SEXUAL ASSAULT']
df['Violent'] = ['Violent' if x in violent else 'Non-Violent' for x in df['Primary Type']]
```

```
df['Violent'].value_counts()
```

```
# label as cannabis or not 
df['Description'].value_counts()
```

```
# clean cannabis discriptions
cannabis = ["MANU/POSS. W/INTENT TO DELIVER: SYNTHETIC MARIJUANA","CALCULATED CANNABIS CONSPIRACY","DELIVER CANNABIS TO PERSON <18","POSSESS - CANNABIS 30 GRAMS OR LESS","MANUFACTURE / DELIVER - CANNABIS OVER 10 GRAM","POSSESS - CANNABIS MORE THAN 30 GRAMS","POSSESSION: SYNTHETIC MARIJUANA","ATTEMPT POSSESSION CANNABIS","CANNABIS PLANT","MANU/DEL:CANNABIS 10GM OR LESS","MANU/DEL:CANNABIS OVER 10 GMS","POSS: CANNABIS 30GMS OR LESS", "POSS: CANNABIS MORE THAN 30GM"]
print(len(cannabis))
cannabis
```

```
df = df.replace('POSSESS - CANNABIS MORE THAN 30 GRAMS','POSS: CANNABIS MORE THAN 30GM')
cannabis.remove('POSSESS - CANNABIS MORE THAN 30 GRAMS')
df = df.replace('POSSESS - CANNABIS 30 GRAMS OR LESS','POSS: CANNABIS 30GMS OR LESS')
cannabis.remove('POSSESS - CANNABIS 30 GRAMS OR LESS')
df = df.replace('POSSESSION: SYNTHETIC MARIJUANA','POSS: SYNTHETIC MARIJUANA')
df = df.replace('MANUFACTURE / DELIVER - CANNABIS OVER 10 GRAM','MANU/DEL:CANNABIS OVER 10 GMS')
cannabis.remove('POSSESSION: SYNTHETIC MARIJUANA')
cannabis.remove('MANUFACTURE / DELIVER - CANNABIS OVER 10 GRAM')
cannabis.append('POSS: SYNTHETIC MARIJUANA')

```

```
print(len(cannabis))
cannabis
```

```
df['Cannabis'] = [True if x in cannabis else False for x in df['Description']]
```

```
df['Cannabis'].value_counts()
```

```
df = df.replace('ATTEMPT ARMED - HANDGUN','POSS: SYNTHETIC MARIJUANA')
df = df.replace('UNLAWFUL USE - HANDGUN','UNLAWFUL USE HANDGUN')
gun.remove('ATTEMPT ARMED - HANDGUN')
gun.remove('UNLAWFUL USE - HANDGUN')
df = df.replace('UNLAWFUL SALE - DELIVERY OF FIREARM AT SCHOOL','UNLAWFUL SALE/DELIVERY OF FIREARM AT SCHOOL')
gun.remove('UNLAWFUL SALE - DELIVERY OF FIREARM AT SCHOOL')
df = df.replace('UNLAWFUL POSSESSION - OTHER FIREARM', 'UNLAWFUL POSS OTHER FIREARM')
gun.remove('UNLAWFUL POSSESSION - OTHER FIREARM')

df = df.replace('UNLAWFUL POSSESSION - HANDGUN', 'UNLAWFUL POSS OF HANDGUN')
gun.remove('UNLAWFUL POSSESSION - HANDGUN')
df = df.replace('UNLAWFUL POSSESSION - AMMUNITION', 'UNLAWFUL POSS AMMUNITION')
gun.remove('UNLAWFUL POSSESSION - AMMUNITION')
df = df.replace('AGGRAVATED PROTECTED EMPLOYEE - OTHER FIREARM','AGG PRO.EMP: OTHER FIREARM')
gun.remove('AGGRAVATED PROTECTED EMPLOYEE - OTHER FIREARM')

df = df.replace('AGGRAVATED PROTECTED EMPLOYEE - HANDGUN','AGG PRO.EMP: HANDGUN')
gun.remove('AGGRAVATED PROTECTED EMPLOYEE - HANDGUN')

df = df.replace('AGGRAVATED - HANDGUN', 'AGGRAVATED: HANDGUN')
gun.remove('AGGRAVATED - HANDGUN')

df = df.replace('AGGRAVATED DOMESTIC BATTERY - HANDGUN', 'AGGRAVATED DOMESTIC BATTERY: HANDGUN')
gun.remove('AGGRAVATED DOMESTIC BATTERY - HANDGUN')

df = df.replace('ARMED - HANDGUN', 'ARMED: HANDGUN')
gun.remove('ARMED - HANDGUN')
df = df.replace('ARMED - OTHER FIREARM','ARMED: OTHER FIREARM')
gun.remove('ARMED - OTHER FIREARM')

df = df.replace('ATTEMPT: ARMED - OTHER FIREARM', 'ATTEMPT: ARMED-OTHER FIREARM')
gun.remove('ATTEMPT ARMED - OTHER FIREARM')

df = df.replace('POSSESS FIREARM / AMMUNITION - NO FOID CARD', 'POSS FIREARM/AMMO:NO FOID CARD')
gun.remove('POSSESS FIREARM / AMMUNITION - NO FOID CARD')
```

```
import datetime
```

```
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')
df.dtypes
```

```
df['Year_Month_Day'] = pd.to_datetime(df['Date']).dt.to_period('D')
```

```
df['Year_Month'] = pd.to_datetime(df['Date']).dt.to_period('M')
```

```
df.dtypes
```

```
df['Month_Num'] = pd.to_datetime(df['Date']).dt.month
```

```
df.head()
```

```
def months(row):
  if row.Month_Num==0:
    month = "January"
  elif row.Month_Num==1:
    month = "February"
  elif row.Month_Num==2:
    month = "March"
  elif row.Month_Num==3:
    month = "April"
  elif row.Month_Num==4:
    month = "May"
  elif row.Month_Num==5:
    month = "June"
  elif row.Month_Num==6:
    month = "July"
  elif row.Month_Num==7:
    month = "August"
  elif row.Month_Num==8:
    month = "September"
  elif row.Month_Num==9:
    month = "October"
  elif row.Month_Num==10:
    month = "November"
  else:
    month = "December"
  return month

df['Month'] = df.apply(months, axis=1)
```

```
df['Month'].value_counts()
```

```
df.head()
```

```
df['Weekday_Num'] = pd.to_datetime(df['Date']).dt.dayofweek
```

```
df['Weekday_Num'].value_counts()
```

```
def weekdays(row):
  if row.Weekday_Num==0:
    day = "Monday"
  elif row.Weekday_Num==1:
    day = "Tuesday"
  elif row.Weekday_Num==2:
    day = "Wednesday"
  elif row.Weekday_Num==3:
    day = "Thursday"
  elif row.Weekday_Num==4:
    day = "Friday"
  elif row.Weekday_Num==5:
    day = "Saturday"
  else:
    day = "Sunday"
  return day 

df['Weekday'] = df.apply(weekdays, axis=1)
```

```
df['Weekday'].value_counts()
```

```
df.head()
```

```
def time_increments(row):
  if row.Date_24h.hour>0 and row.Date_24h.hour<4:
    val = "[00:00-3:59]"
  elif row.Date_24h.hour<8:
    val = "[04:00-7:59]"
  elif row.Date_24h.hour<12:
    val = "[08:00-11:59]"
  elif row.Date_24h.hour<16:
    val = "[12:00-15:59]"
  elif row.Date_24h.hour<20:
    val = "[16:00-19:59]"
  else:
    val = "[20:00-23:59]"
  return val 

df['Time_Bucket'] = df.apply(time_increments, axis=1)
```
